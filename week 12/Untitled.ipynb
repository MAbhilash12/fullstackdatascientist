```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
print(iris)
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
X=pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width','petal length','petal width'])
```

```python
X
```

```python
y=iris['target']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(max_depth=3)
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred=dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print("accuracy score:",accuracy_score(y_test,y_pred))
print("classification report:\n",classification_report(y_test,y_pred))
```

# DT Classifier with prepuning and Hyperparameter Tuning


```python
param={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','raandom'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel=DecisionTreeClassifier()
```

```python
GRID=GridSearchCV(treemodel,param_grid=param,cv=5,scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
GRID.fit(X_train,y_train)
```

```python
GRID.best_params_
```

```python
y_pred1=GRID.predict(X_test)
```

```python
y_pred1
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred1)
```

```python
cm
```

```python
print(classification_report(y_test,y_pred1))
```
