```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False, drop='first')

categorical_cols = ['city_tier', 'department', 'education_level']

ohe_df = pd.DataFrame(
    ohe.fit_transform(df[categorical_cols]),
    columns=ohe.get_feature_names_out(categorical_cols)
)

df = pd.concat([df.drop(columns=categorical_cols), ohe_df], axis=1)

df.head()
```

```python
df
```

```python
x=df.drop(['target_left_company'],axis=1)
x.head()
```

```python
y=df['target_left_company']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier(n_estimators=100,random_state=42)
rf_model.fit(x_train,y_train)
```

```python
y_pred=rf_model.predict(x_test)
```

```python
comparison_df = pd.DataFrame({'Actual': y_test,'Predicted': y_pred})
comparison_df
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
print("accuracy score: ",accuracy_score(y_test,y_pred))
print("classification report:\n ",classification_report(y_test,y_pred))
print("confusion matrix: \n",confusion_matrix(y_test,y_pred))
```
