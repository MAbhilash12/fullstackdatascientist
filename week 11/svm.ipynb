```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=2, n_classes=2,n_informative=2,n_redundant=0,random_state=42)
```

```python
x
```

```python
y
```

```python
pd.DataFrame(x)[0]
```

```python
sns.scatterplot(
    x=pd.DataFrame(x)[0],
    y=pd.DataFrame(x)[1],
    hue=y
)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(x_train,y_train)
```

```python
svc.coef_
```

```python
y_pred=svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
rbf=SVC(kernel='rbf')
```

```python
rbf.fit(x_train,y_train)
```

```python
y_pred1=rbf.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred1))
```

```python
print(classification_report(y_test,y_pred1))
```

```python
polynom=SVC(kernel='poly')
```

```python
polynom.fit(x_train,y_train)
y_pred2=polynom.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
```

```python
sigmo=SVC(kernel='sigmoid')
sigmo.fit(x_train,y_train)
y_pred3=sigmo.predict(x_test)
print(confusion_matrix(y_test,y_pred3))
print(classification_report(y_test,y_pred3))
```

```python
#Hyper parameter Tuning with SVC
from sklearn.model_selection import GridSearchCV

#defining parameters range 
param_grid={'C':[0.1,1,10,100,1000],
            'gamma':[1,0.1,0.01,0.001,0.0001],
            'kernel':['rbf']}
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_params_

y_pred4=grid.predict(x_test)

print(classification_report(y_test,y_pred4))

print(confusion_matrix(y_test,y_pred4))


```


C = Regulation Parameter

gamma=Influence of a single data point

Total Combinations: 5(C)*5(gamma)*1(kernel)=25 models

Here we used GridSearchCV is used to perform hyperparmeter tunning.

It check all combos of given parameters using cv and selects the model with best performance....
